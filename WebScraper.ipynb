```python
from selenium import webdriver

from selenium.webdriver.chrome.service import Service

from selenium.webdriver.support.ui import WebDriverWait

from selenium.webdriver.support import expected_conditions as EC

from bs4 import BeautifulSoup as bs

import codecs

import re

from webdriver_manager.chrome import ChromeDriverManager

import requests

from selenium.webdriver.common.by import By

import time

import pandas as pd
```

```python
def getMoreInfo(driver,vehicle):
    
    #Getting the new URL after clicking the first element
#     driver.get(driver.current_url)
#     driver.implicitly_wait(5)
#     time.sleep(3)
    print("yes")
    
    # Clicking the Showmore button
    try:
        driver.find_element(By.CLASS_NAME,"showMoreButton-2571466895").click()
        driver.implicitly_wait(5)
    except:
        pass
   
    
    try:
        condition = driver.find_element(By.XPATH,"//dd[contains(@itemprop, 'itemCondition')]").text
    except :
        condition = None
    
    try:
        year = driver.find_element(By.XPATH,"//dd[contains(@itemprop, 'vehicleModelDate')]").text
    except :
        year = None
        
    try:
        make = driver.find_element(By.XPATH,"//dd[contains(@itemprop, 'brand')]").text
    except :
        make = None
        
    try:
        model = driver.find_element(By.XPATH,"//dd[contains(@itemprop, 'model')]").text
    except :
        model = None
    
    try:
        trim = driver.find_element(By.XPATH,"//dd[contains(@itemprop, 'vehicleConfiguration')]").text
    except :
        trim = None
    
    try:
        color = driver.find_element(By.XPATH,"//dd[contains(@itemprop, 'color')]").text
    except :
        color = None
    
    try:
        bodyType = driver.find_element(By.XPATH,"//dd[contains(@itemprop, 'bodyType')]").text
    except :
        bodyType = None
        
    try:
        numberOfDoors = driver.find_element(By.XPATH,"//dd[contains(@itemprop, 'numberOfDoors')]").text
    except :
        numberOfDoors = None
        
    try:
        seats = driver.find_element(By.XPATH,"//dd[contains(@itemprop, 'seatingCapacity')]").text
    except :
        seats = None
        
    try:
        driveTrain = driver.find_element(By.XPATH,"//dd[contains(@itemprop, 'driveWheelConfiguration')]").text
    except :
        driveTrain = None
        
    try:
        transmission = driver.find_element(By.XPATH,"//dd[contains(@itemprop, 'vehicleTransmission')]").text
    except :
        transmission = None
        
    try:
        mileage = driver.find_element(By.XPATH,"//dd[contains(@itemprop, 'mileageFromOdometer')]").text
    except :
        mileage = None
    
    vehicle["Condition"] = condition
    vehicle["year"] = year
    vehicle["Make"] = make
    vehicle["Model"] = model
    vehicle["Trim"] = trim
    vehicle["Color"] = color
    vehicle["Body Type"] = bodyType
    vehicle["Number of Doors"] = numberOfDoors
    vehicle["Number of Seats"] = seats
    vehicle["Drive Train"] = driveTrain
    vehicle["Transmission"] = transmission
    vehicle["Mileage"] = mileage
    
    
```

```python
def loadingWebDriver():
    driver=webdriver.Chrome(service=Service(ChromeDriverManager().install()))
    return driver
```

```python
def getUrl(driver,url):
    driver.get(url)
    time.sleep(2)
```

```python
def scrapingPage(driver):
    
#     vehiclesInfos2 = driver.find_elements(By.CLASS_NAME,'info')
    time.sleep(5)
#     runLength = len(vehiclesInfos1)
    runLength = 30

    for i in range(runLength):

        driver.get(driver.current_url)

        vehiclesInfos2 = driver.find_elements(By.CLASS_NAME,'info')
#         driver.implicitly_wait(5)

        print(i," ",len(vehiclesInfos2))

        carName = vehiclesInfos2[i].find_element(By.CLASS_NAME,'title')
        price = vehiclesInfos2[i].find_element(By.CLASS_NAME,'price')
#         distance = vehiclesInfos2[i].find_element(By.CLASS_NAME,'distance')
        location = vehiclesInfos2[i].find_element(By.CLASS_NAME,'location')
        description = vehiclesInfos2[i].find_element(By.CLASS_NAME,'description')

        vehicle = {}
        vehicle["Title"] = carName.text
        vehicle["Price"] = price.text
#         vehicle["Distance"] = distance.text
        vehicle["Location"] = location.text
        vehicle["Description"] = description.text

    #     Clicking the Nameof the vehicle to load the next page

        try:
            driver.find_element(By.LINK_TEXT,str(carName.text)).click()
#             driver.implicitly_wait(10)
            time.sleep(3)
        except:
            pass

        getMoreInfo(driver,vehicle)

        vehicles.append(vehicle)
        
#         time.sleep(2)

        driver.back()

        time.sleep(3)

    #     driver.implicitly_wait(25)
    
```

```python
def myScraper():
    
    driver = loadingWebDriver()
    
    number_of_pages = 100
    j = 0 
#     vehicles = []
    
    
    while j <= number_of_pages:
        print("Page Number:", j+1)
        if j == 0:
            url = "https://www.kijiji.ca/b-cars-trucks/gta-greater-toronto-area/new__used/c174l1700272a49"
#             print(url)
        else:
            url = f"https://www.kijiji.ca/b-cars-trucks/gta-greater-toronto-area/new__used/page-{j+1}/c174l1700272a49"
#             print(url)
        
        getUrl(driver,url)
        
        scrapingPage(driver)
        
        j+=1
    
```
