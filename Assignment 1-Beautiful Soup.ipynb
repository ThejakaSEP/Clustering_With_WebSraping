```python
import numpy as np
import pandas as pd

import requests
from bs4 import BeautifulSoup as bs
```

```python
URL = "https://www.kijijiautos.ca/cars/#od=down&sb=relv2"
page = requests.get(URL)
print(page.text)
```

```python
print(page.content)
```

```python
soup = bs(page.text, "lxml")
print(soup.prettify())
```

```python
ads = soup.find_all("div", attrs={"class":"ccN7dZ"})
```

```python
vehicles = []
for ad in ads:
    vehicle = {}
#     a1 = ad.find("div",attrs={"class":"ecN7dZ"})
    
    #Model Name
    model = ad.find("h2",attrs={"class":"G2jAym fcN7dZ z2jAym p2jAym b2jAym"})
    
    #Getting the Attributes
    attributes = ad.find_all("li",attrs={"class":"c3n44x e3n44x"})
    
    vehicle["Model"] = model.text
    
    try:
        mileage = attributes[0].find("span",attrs={"class":"G2jAym y2jAym q2jAym"})
        vehicle["Mileage"] = mileage.text
    except IndexError:
        vehicle["Mileage"] = None
    
    try:
        location = attributes[1].find("span",attrs={"class":"G2jAym y2jAym q2jAym"})
        vehicle["Location"] = location.text
    except IndexError:
        vehicle["Location"] = None
    
    try:
        transmission = attributes[2].find("span",attrs={"class":"G2jAym y2jAym q2jAym"})
        vehicle["Transmission"] = transmission.text
    except IndexError:
        vehicle["Transmission"] = None
        
    
    F = ad.find_all("li",attrs={"class":"c3n44x"})
    fuel = F[3].find("span",attrs={"class":"G2jAym y2jAym q2jAym"})
    vehicle["Fuel"] = fuel.text
    
    vehicles.append(vehicle)

    
```
