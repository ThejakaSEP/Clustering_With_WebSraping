```python
from selenium import webdriver

from selenium.webdriver.chrome.service import Service

from selenium.webdriver.support.ui import WebDriverWait

from selenium.webdriver.support import expected_conditions as EC

from bs4 import BeautifulSoup as bs

import codecs

import re

from webdriver_manager.chrome import ChromeDriverManager

import requests

from selenium.webdriver.common.by import By
```

```python
driver=webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
val = input("Enter a url:")

wait = WebDriverWait(driver, 10)

driver.get(val)


get_url = driver.current_url
wait.until(EC.url_to_be(val))


if get_url == val:
    page_source = driver.page_source
```

```python
page = requests.get('https://www.kijiji.ca/b-cars-trucks/gta-greater-toronto-area/new__used/c174l1700272a49')
```

```python
soup = bs(page.text, "lxml")
```

```python
ads = soup.find_all("div", attrs={"class":"info"})
```

```python
len(ads)
```

```python
vehicles = []
for ad in ads:
    vehicle = {}
    vehicle["Title"] = ad.find("div",attrs={"class":"title"}).text.strip()
    vehicle["Price"] = ad.find("div",attrs={"class":"price"}).text.strip()
    vehicle["Distance"] = ad.find("div",attrs={"class":"distance"}).text.strip()
    vehicle["Location"] = ad.find("div",attrs={"class":"location"}).text.strip()
    vehicle["Description"] = ad.find("div",attrs={"class":"description"}).text.strip()
    
    title = ad.find("div",attrs={"class":"title"}).text.strip()

    driver.find_element(By.LINK_TEXT,title).click()
    driver.implicitly_wait(15)
    driver.back() 
    driver.implicitly_wait(15)
    
    vehicles.append(vehicle)
print(title)
```
